L'objectif de ce notebook est de tenter de determiner statistiquement la raison qu'un joueur soit drafté

On sépare en deux : les joueurs non draftés en fin de carrière universitaires des joueurs draftés

```python
df_drafted = df.loc[df["drafted"] == 1]
df_non_drafted = df.loc[df["drafted"] == 0]
```

## Conférences

### Divisions

La Division I (DI), composée de 32 conférences, est le plus haut niveau de sport universitaire de la National Collegiate Athletic Association (ou NCAA) aux États-Unis. Les universités qui en sont membres sont généralement réputées pour leurs athlètes. Les niveaux inférieurs à la DI sont la DII et la DIII. La séparation entre ces deux dernières divisions se fait de la façon suivante :
- la Division 2 procurent des bourses 
- la Division 3 ne procurent pas de bourses

La globalité des joueurs draftés ont joué en première division. Le dataset scrappé ne possède que des universitaires de cette division.

### Rôle des joueurs

![title](resource/Basketball_Positions.png)  

En basketball, il existe 5 rôles de joueurs. Cependant, il est difficile parfois de déterminer précisément le rôle de certains joueurs. Dans le cas du dataset scrappé, on catégorise les joueurs en trois types : 
- Guard (Point guard and shooting guard) : qui ont pour rôle d'etre les meneurs de jeu
- Forward (Small Forward et Power Forward) : qui ont pour rôle de marquer le plus de points
- Center (Center) : le plus grand de taille de l'équipe qui joue à proximité du panier

```python
def players_by_position(df,pos):
    return df.loc[df["position"] == pos]
```

```python
# Les joueurs par positions
sns.set(rc={'figure.figsize':(12,8)})
labels = 'Guard', 'Forward', 'Center'
sizes = [players_by_position(df,x).shape[0] for x in labels]
explode = (0, 0, 0) 

fig1, ax1 = plt.subplots()
plt.title("Répartition des postes en NCAA")
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True)
ax1.axis('equal');
```

```python
sizes = [players_by_position(df_drafted,x).shape[0] for x in labels]
explode = (0, 0, 0) 

fig1, ax2 = plt.subplots()
ax2.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True)
ax2.axis('equal');
```

On remarque qu'en NCAA, il y a plus de "guards" que de "forward" ou "center" mais les équipes NBA
ont plus tendance à drafter des forwards.  
Serait-il une question de taille ?

```python
def comparer_ncaa_drafted(category) :
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=3,  figsize=(18,16))
    fig.suptitle(f'Répartition des {category} par position', fontsize = 15)

    players = ['ncaa', 'drafted']

    ax1.hist(df.loc[df["position"] == 'Guard',category], density = True)
    ax2.hist(df.loc[df["position"] == 'Forward',category], density = True)
    ax3.hist(df.loc[df["position"] == 'Center',category], density = True)

    ax1.hist(df_drafted.loc[df_drafted["position"] == 'Guard', category], density = True, alpha = 0.5)
    ax2.hist(df_drafted.loc[df_drafted["position"] == 'Forward',category], density = True, alpha = 0.5)
    ax3.hist(df_drafted.loc[df_drafted["position"] == 'Center',category], density = True, alpha = 0.5)

    fig.text(0.5, 0.04, category, ha='center', va='center')

    ax1.set_title('Guard')
    ax2.set_title('Forward')
    ax3.set_title('Center')


    plt.show()
```

```python
comparer_ncaa_drafted("height")
```

On peut déduire que les joueurs draftés ont tendance à être plus grand que la moyenne. 
On peut conclure que la taille d'un joueur est un facteur déterminant pour passer professionnel.

### Statistiques de bases  

Les feuilles de matchs indiquent les statistiques en valeur absolue pour chacune des cinq catégories principales : points, rebonds, passes décisives, interceptions et contres.  


Vérifions si ces statistiques basiques impactent la selection à la draft

```python
sns.pairplot(df[["points",'total_rebounds', "assists", "steals","blocks","drafted"]],corner = True, hue = "drafted");
```

La fonction pairplot de Seaborn permet d'afficher les graphiques de densités et de déduire la colinéarité entre les statistiques.  
On remarque que les joueurs draftés ont tendance à avoir de meilleurs performances à chaque match que les non draftés.  
Pairplot ne nous permet pas de déduire une relation entre les statistiques

```python
df_encoding = df.copy()
from sklearn.preprocessing import LabelEncoder
df_encoding["conference"] = LabelEncoder().fit_transform(df_encoding['conference'])
df_encoding["team_abbreviation"] = LabelEncoder().fit_transform(df_encoding['team_abbreviation'])
df_encoding["position"] = LabelEncoder().fit_transform(df_encoding['position'])
```

```python
corr_df = df_encoding[["points",'total_rebounds', "assists", "steals","blocks","height","position","conference","team_abbreviation","drafted"]].corr()
mask_ut=np.triu(np.ones_like(corr_df, dtype=bool))
sns.heatmap(corr_df,annot=True, cmap="Spectral", mask=mask_ut);
```
