### Rôle des joueurs

![title](resource/Basketball_Positions.png)  

En basketball, il existe 5 rôles de joueurs. Cependant, il est difficile parfois de déterminer précisément le rôle de certains joueurs. Dans le cas du dataset scrappé, on catégorise les joueurs en trois types : 
- Guard (Point guard and shooting guard) : qui ont pour rôle d'etre les meneurs de jeu
- Forward (Small Forward et Power Forward) : qui ont pour rôle de marquer le plus de points
- Center (Center) : le plus grand de taille de l'équipe qui joue à proximité du panier

```python
def players_by_position(df,pos):
    return df.loc[df["position"] == pos]
```

```python
# Les joueurs par positions
sns.set(rc={'figure.figsize':(12,8)})
labels = 'Guard', 'Forward', 'Center'
sizes = [players_by_position(df,x).shape[0] for x in labels]
explode = (0, 0, 0) 

fig1, ax1 = plt.subplots()
plt.title("Répartition des postes en NCAA")
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True)
ax1.axis('equal');
```

```python
sizes = [players_by_position(df_drafted,x).shape[0] for x in labels]
explode = (0, 0, 0) 

fig1, ax2 = plt.subplots()
ax2.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True)
ax2.axis('equal');
```

On remarque qu'en NCAA, il y a plus de "guards" que de "forward" ou "center" mais les équipes NBA
ont plus tendance à drafter des forwards.  
Serait-il une question de taille ?

```python
def comparer_ncaa_drafted(category) :
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=3,  figsize=(18,16))
    fig.suptitle(f'Répartition des {category} par position', fontsize = 15)

    players = ['ncaa', 'drafted']

    ax1.hist(df.loc[df["position"] == 'Guard',category], density = True)
    ax2.hist(df.loc[df["position"] == 'Forward',category], density = True)
    ax3.hist(df.loc[df["position"] == 'Center',category], density = True)

    ax1.hist(df_drafted.loc[df_drafted["position"] == 'Guard', category], density = True, alpha = 0.5)
    ax2.hist(df_drafted.loc[df_drafted["position"] == 'Forward',category], density = True, alpha = 0.5)
    ax3.hist(df_drafted.loc[df_drafted["position"] == 'Center',category], density = True, alpha = 0.5)

    fig.text(0.5, 0.04, category, ha='center', va='center')

    ax1.set_title('Guard')
    ax2.set_title('Forward')
    ax3.set_title('Center')


    plt.show()
```

```python
comparer_ncaa_drafted("height")
```

On peut déduire que les joueurs draftés ont tendance à être plus grand que la moyenne. 
On peut conclure que la taille d'un joueur est un facteur déterminant pour passer professionnel.

### Poids

```python
comparer_ncaa_drafted("weight")
```

Le poids d'un joueur ne semble pas être un paramètre

## Statistiques de bases  

Les feuilles de matchs indiquent les statistiques en valeur absolue pour chacune des cinq catégories principales : points, rebonds, passes décisives, interceptions et contres.  


Vérifions si ces statistiques basiques impactent la selection à la draft

```python
sns.pairplot(df[["points",'total_rebounds', "assists", "steals","blocks","drafted"]],corner = True, hue = "drafted");
```

La fonction pairplot de Seaborn permet d'afficher les graphiques de densités et de déduire la colinéarité entre les statistiques.  
On remarque que les joueurs draftés ont tendance à avoir de meilleurs performances à chaque match que les non draftés.  
Pairplot ne nous permet pas de déduire une relation entre les statistiques

```python
df_encoding = df.copy()
from sklearn.preprocessing import LabelEncoder
df_encoding["conference"] = LabelEncoder().fit_transform(df_encoding['conference'])
df_encoding["team_abbreviation"] = LabelEncoder().fit_transform(df_encoding['team_abbreviation'])
df_encoding["position"] = LabelEncoder().fit_transform(df_encoding['position'])
```

```python
corr_df = df_encoding[["points",'total_rebounds', "assists", "steals","blocks","height","position","conference","team_abbreviation","drafted"]].corr()
mask_ut=np.triu(np.ones_like(corr_df, dtype=bool))
sns.heatmap(corr_df,annot=True, cmap="Spectral", mask=mask_ut);
```

On retrouve dans cette matrice de corrélation, certaines remarques ont pu être justifiées :
- il y a une forte colinéarité entre la taille et la position
- les meneurs de jeu ont tendance à réaliser beaucoup de passes décisives et d'interception (d'où leur forte corrélation)

Cependant, aucune feature possède une grande corrélation avec le label "drafté". La corrélation points/drafté n'est que de 0.25, on aurait pensé qu'un joueur marquant beaucoup de points serait un facteur determinant pour sa selection à la draft mais cette statistiques ne dévoilent pas l'**efficacité** du joueur :  

Un joueur marquant 15 points avec une efficacité de 85% (tirs rentrés/tirs tentés) a plus de chance d'être selectionné qu'un joueur marquant 30 points avec une efficacité de 45%.
Les tirs ratés sont des balles perdues pour l'équipe qui pourraient encaisser des points en contre attaque.  

De ce fait de nouvelles métriques sont utilisés pour mieux juger un joueur : les **statistiques avancées**.

## Statistiques Avancées

### Statistiques de tirs

![title](resource/court.jpg)  

Au basketball, il existe 3 types de tirs : 
- un **lancer franc (1 point)** est une pénalité accordée à un joueur victime d'une faute au moment où il tirait.
- un **"two point shot" (2 points)** est un tir tenté en match à l'intérieur de la "three-point line" et la "boundary line"
- un **"three point shoot" (3 points)** est un tir tenté en match à l'extérieur de la "three-point line" et la "boundary line"

L'efficacité de ces trois types de tirs sont determinés par le rapport $\frac{Tirs.mis}{Tirs.tentés}$

Cependant, ces tirs n'ont pas la même efficacité mais un tir à 3 points rapporte plus que de points qu'un lancer franc. De ce fait, les statisticiens ont créé une nouvelles statistiques qui est une moyenne pondérée des 3 types de tirs :   

$TS\% = \frac{PTS}{2(FGA + (0.44 \times FTA))} \times 100 \% $



```python
sns.pairplot(df[["effective_field_goal_percentage","field_goal_percentage", "free_throw_percentage","three_point_percentage","two_point_percentage","true_shooting_percentage","drafted"]],corner = True ,hue="drafted");
```

On remarque que les joueurs draftés possèdent de meilleurs statistiques avancées pour le tir.  
Maintenant vérifions les statistiques avancées pour les autres statistiques de base

## Pour aller  plus loin : Classement des joueurs NCAA lors de leurs dernières de lycées

Aux Etats Unis, les joueurs même encore au lycée sont supervisés par des journalistes et des recruteurs professionnels. A chaque fin de saison, un classement des 100 meilleurs "prospects" du pays est réalisé.  

On s'intéresse au classement aux joueurs ayant terminé le lycée. Les joueurs les mieux classés recoivent des bourses scolaires des meilleurs universités du pays donc recoivent une formation de basket-ball de haut niveau.

```python
df_rsci = pd.read_csv("data/rsci.csv",  index_col=0)
df_rsci
```

```python
df_new = pd.merge(df_encoding,df_rsci,left_on='player_id',right_on='id',how='left')
df_new = df_new.drop(columns=['id'])
df_new["Ranking"] = df_new['Ranking'].fillna(-100)
```

```python
df_new["player_id"] = LabelEncoder().fit_transform(df_new['player_id'])
```

```python
sns.scatterplot(data=df_new, x="player_id", y="Ranking" ,hue="drafted");
```

Les joueurs ayant un ranking de "-100" représentent les joueurs qui n'étaient pas classés dans le top 100 au lycée.

```python
sns.heatmap(df_new[["Ranking","drafted"]].corr(), annot = True, cmap="Spectral")
```

```python
prc_rsci_ncaa = df_new[df_new["drafted"] == 1]["drafted"].count()/df_new[df_new["Ranking"] > 0]["Ranking"].count()
f"Au final, {round(prc_rsci_ncaa*100,2)}% des 100 meilleurs lycéens finissent par devenir professionnels"
```
