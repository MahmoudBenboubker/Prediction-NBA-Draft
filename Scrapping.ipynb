```python
from sportsreference.ncaab.roster import Roster
purdue = Roster('PURDUE')
players_id = []
for player in purdue.players:
    # Prints the name of all players who played for Purdue in the most
    # recent season.
    players_id.append(player.player_id)
```

```python
def players_info(id):
    from sportsreference.ncaab.roster import Player
    player = Player(id)
    df_p = player.dataframe
    df_p["name"] = player.name
    df_p["id"] = player.player_id
    return df_p
    # Prints a Pandas DataFrame of all relevant stats per season for Edwards
```

```python
players_dataframes = []
for id in players_id :
    players_dataframes.append(players_info(id))
```

```python
def change_position_col(col_name, df):
    col = df.pop(col_name)
    df.insert(0,col.name,col)
```

```python
df_all_players = pd.concat(players_dataframes)
change_position_col("name", df_all_players)
change_position_col("id", df_all_players)
df_all_players
```

```python
def clean_df(csv_file,i):
    df = pd.read_csv(csv_file)
    df.dropna(subset = ["College"], inplace=True)
    df["Player"] = df["Player"].apply(lambda x : x.split("\\")[0])
    df["index"] = df["Pk"].apply(lambda x : str(i)+"/" + str(x))
    df.set_index("index")
    return df
```

```python
%%time
df_draftclasses = []
for i in range(2010,2020):
    df_draftclasses.append(clean_df(f'rookie-{i}.csv',i))
df_draftclasses = pd.concat(df_draftclasses)
```

```python
df_draftclasses.set_index('index').to_csv("all_rookies.csv")
```

```python
players = df["Player"].values    
players = [player.lower().replace(' ','-').replace('.','') + '-1' for player in players]
players
```

```python
players[19] = 'bj-mullens-1'
players[24] = 'jeff-pendergraph-1'
players.remove('patty-mills-1')
```
